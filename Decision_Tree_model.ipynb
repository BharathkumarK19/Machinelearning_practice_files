Decision Tree Model

```python
#Decision Trees
#import libraries


import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import accuracy_score
```

```python
iris=load_iris()
```

```python

iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target # Add the target variable to the DataFrame
```

```python
# define input X and output Y variables
x=iris_df.iloc[:,:4]
y=iris_df.iloc[:,-1]
```

```python
#Splitting data for the training and the testing
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=4519)
```

```python
#Standardizing the input variables
scaler=StandardScaler()
scaler.fit_transform(x_train)
scaler.fit_transform(x_test)
```

```python
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
plot_tree(model)
```

```python
#Finding the accuracy of model

accuracy_score(y_test,y_pred)
```
